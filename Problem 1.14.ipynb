A random variable $X$ has a probability density function defined as follows:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 3, 100)

# Probability Density Function
y1 = np.piecewise(x, [x<0, (x>=0) & (x<=2), x>2], [lambda x: 0, lambda x: 1/2*x, lambda x: 0])
plt.plot(x, y1)
plt.title("Probability Density Function")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
# Cumulative Distribution Function
y2 = np.piecewise(x, [x<0, (x>=0) & (x<=2), x>2], [lambda x: 0, lambda x: 1/4*x**2, lambda x: 1])

plt.plot(x, y2)
plt.title("Cumulative Density Function")
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
EX = np.sum(x*y1)*(x[1]-x[0])
EX2 =np.sum(x**2*y1)*(x[1]-x[0])

print("E(X) = %.1f" % (EX))
print("E(X2) = %.1f" % (EX2))
print("Var(X) = %.1f" % (EX2-(EX)**2))
```
