Noise measurements of a certain amplifier indicate that the rms output voltage is 100 uV. The frequency spectrum of the noise is constant from 0 to 10MHz and zero otherwise. (Hint: example 3.9)

1. Determine the spectral density function and the autocorrelation function

```python
from numpy import linspace, piecewise, sin, pi
import matplotlib.pyplot as plt

w = linspace(-20, 20, 100)
tau = linspace(-20, 20, 100)

# Probability Density Function
y1 = piecewise(w, [w < -10, (w >= -10) & (w <= 10), w > 10], [lambda w: 0, lambda w: 1/20 , lambda w: 0])

# Autocorrelation
y2 = sin(2*pi*10*tau)/(2*pi*10*tau)

# Plot results
plt.plot(w, y1)
plt.title("Spectral Density Function")
plt.xlabel("MHz")
plt.ylabel("uV^2/MHz")
plt.grid()
plt.show()

plt.plot(tau, y2)
plt.title("Autocorrelation Function")
plt.xlabel("1e-8 second")
plt.ylabel("uV^2")
plt.grid()
plt.show()
```
